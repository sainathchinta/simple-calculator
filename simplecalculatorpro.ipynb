```python
from tkinter import *


#functions for buttons


def press(number):
    curr=entrybar.get()
    entrybar.delete(0,END)
    entrybar.insert(0,str(curr)+str(number))
def clear():
    entrybar.delete(0,END)


def add():
    firstno=entrybar.get()
    global fnum
    global math
    math="add"
    fnum=int(firstno)
    entrybar.delete(0,END)
    
def sub():
    firstno=entrybar.get()
    global fnum
    global math
    math="sub"
    fnum=int(firstno)
    entrybar.delete(0,END)
    
def mul():
    firstno=entrybar.get()
    global fnum
    global math
    math="mul"
    fnum=int(firstno)
    entrybar.delete(0,END)

def div():
    firstno=entrybar.get()
    global fnum
    global math
    math="div"
    fnum=int(firstno)
    entrybar.delete(0,END)

def equal():
    secondno=entrybar.get()
    entrybar.delete(0,END)
    
    if math=="add":
        entrybar.insert(0,fnum+int(secondno))
    
    if math=="sub":
        entrybar.insert(0,fnum-int(secondno))
    
    if math=="mul":
        entrybar.insert(0,fnum*int(secondno))
        
    if math=="div":
        entrybar.insert(0,fnum/int(secondno))

#mainwindow        

root=Tk()

root.title("simple calculator")
root.geometry("270x200")
entrybar=Entry(root)
entrybar.grid(columnspan=4,ipadx=70)


#creation of buttons

button1=Button(root,text="1",command=lambda:press(1),fg="black",bg="white",height=1,width=7)
button2=Button(root,text="2",command=lambda:press(2),fg="black",bg="white",height=1,width=7)
button3=Button(root,text="3",command=lambda:press(3),fg="black",bg="white",height=1,width=7)

button4=Button(root,text="4",command=lambda:press(4),fg="black",bg="white",height=1,width=7)
button5=Button(root,text="5",command=lambda:press(5),fg="black",bg="white",height=1,width=7)
button6=Button(root,text="6",command=lambda:press(6),fg="black",bg="white",height=1,width=7)

button7=Button(root,text="7",command=lambda:press(7),fg="black",bg="white",height=1,width=7)
button8=Button(root,text="8",command=lambda:press(8),fg="black",bg="white",height=1,width=7)
button9=Button(root,text="9",command=lambda:press(9),fg="black",bg="white",height=1,width=7)

button0=Button(root,text="0",command=lambda:press(0),fg="black",bg="white",height=1,width=7)
button11=Button(root,text="=",command=equal,fg="black",bg="white",height=1,width=7)
button12=Button(root,text="+",command=add,fg="black",bg="white",height=1,width=7)
button13=Button(root,text="-",command=sub,fg="black",bg="white",height=1,width=7)
button14=Button(root,text="*",command=mul,fg="black",bg="white",height=1,width=7)
button15=Button(root,text="/",command=div,fg="black",bg="white",height=1,width=7)
button16=Button(root,text="clear",command=clear,fg="black",bg="white",height=1,width=7)

#grid for buttons

button1.grid(row=1,column=0)
button2.grid(row=1,column=1)
button3.grid(row=1,column=2)

button4.grid(row=2,column=0)
button5.grid(row=2,column=1)
button6.grid(row=2,column=2)

button7.grid(row=3,column=0)
button8.grid(row=3,column=1)
button9.grid(row=3,column=2)

button0.grid(row=5, column=0)
button11.grid(row=5, column=1)
button12.grid(row=5, column=2)
button13.grid(row=6, column=0)
button14.grid(row=6, column=1)
button15.grid(row=6, column=2)

button16.grid(row=7,column=0)

root.mainloop()

```
